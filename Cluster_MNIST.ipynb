```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
from sklearn.manifold import TSNE
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data')
import tensorflow as tf

```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
#separate the data into 10 groups by digits.
cluster={}
for i in range(10):
    cluster[i]=mnist.train.images[mnist.train.labels==i]
```

```python
#define a function to do the DBSCAN and Kmeans cluster, eps1 is the eps of DBSCAN, min_samples1 is the minimum 
#density of DBSCAN, j was the digit we want to cluster. The cluster number of Kmeans is the same as that of DBSCAN.
def Cluster_MNIST(eps1,min_samples1,j):
    DBSCAN_model = DBSCAN(eps = eps1, min_samples =min_samples1).fit_predict(cluster[j])
    kmeans_model = KMeans(n_clusters=len(set(DBSCAN_model))).fit(cluster[j])
    fig = plt.figure(figsize=(12,10))
    print('cluster '+str(j))
    print('Centers of DBSCAN clusters')
    for i in set(DBSCAN_model):
        ax = fig.add_subplot(4, 5, i+2, xticks=[], yticks=[]) 
        x=cluster[j][DBSCAN_model==i]
        plt.imshow((sum(x)/x.shape[0]).reshape(28,28))
    plt.show()
    print('--------------------------------------------------------')
    print('Centers of Kmeans clusters')
    fig2 = plt.figure(figsize=(12,10))
    for i in range(len(set(DBSCAN_model))):
        ax2 = fig2.add_subplot(4, 5, i+1, xticks=[], yticks=[]) 
        plt.imshow(kmeans_model.cluster_centers_[i].reshape(28,28))
    plt.show()
```
